# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
Base.classes.measurement.__table__.columns.keys()
```

```python
last_data = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
last_data
```

```python
date_in = dt.datetime(2016, 8, 23)
last_12prcp = session.query(Measurement.prcp, Measurement.date).filter(Measurement.date > date_in).order_by(Measurement.date.asc()).all()

last_12prcp
```

```python
prcp_date = pd.DataFrame(last_12prcp, columns = ["prcp", "date"])
prcp_date.set_index("date", inplace =True)
prcp_date
```

```python
prcp_date.plot()
plt.tight_layout()
plt.show()
```

```python
# Design a query to show how many stations are available in this dataset?
```

```python
station_active = session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).\
order_by(func.count(Measurement.station).desc()).all()
station_active
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
```

```python
temp_high = session.query(func.min(Measurement.tobs),func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
filter(Measurement.station == 'USC00519281').all()
temp_high
```
